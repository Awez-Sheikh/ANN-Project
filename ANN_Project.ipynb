```python
# Load the .mat file
data = loadmat("Matlab_cancer.mat")

# Extract features and labels
X = data['x']
y = data['t']
# Convert to DataFrame for easier manipulation
X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

```python
# Transpose DataFrame y
y_transposed = y.T
print("\nTransposed DataFrame y:")
print(y_transposed)
```

```python
# Ensure indices match (usually true by default)
assert len(X_transposed) == len(y_transposed), "Length of X and y must match"
```

```python
dfy=pd.DataFrame(y_transposed)
dfy
```

```python
# Define a mapping function to convert binary pair to label
def map_binary_to_label(row):
    if row[0] == 0 and row[1] == 1:
        return 'Normal'
    elif row[0] == 1 and row[1] == 0:
        return 'Cancer'
    else:
        return 'Unknown'  # In case there are any other combinations

# Apply the function to each row and create a new column
dfy['N/C'] = dfy.apply(map_binary_to_label, axis=1)

print("\nDataFrame with Target Labels:")
print(dfy[['N/C']])
```
